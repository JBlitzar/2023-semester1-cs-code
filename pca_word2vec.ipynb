```python
import gensim
from nltk.data import find

word2vec_sample = str(find('models/word2vec_sample/pruned.word2vec.txt'))
model = gensim.models.KeyedVectors.load_word2vec_format(word2vec_sample, binary=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
max_count = 100
labels = []
count = 0
model_dim = len(model["university"])
X = np.zeros(shape=(max_count,model_dim))
for term in model.index_to_key:
    X[count] = model[term]
    labels.append(term)
    count+= 1
    if count >= max_count:
        break
```

```python
def PCA(D_prime, X, D):
    N=len(X)
    X = (X-np.mean(X,axis=0))/(np.std(X,axis=0)) # step 1
    print(X.shape)
    covariance = (np.matmul(X.transpose(), X))/(N-1) # step 2
    print(covariance.shape)

    eigvals, eigvecs = np.linalg.eig(covariance) # step 3
    eigenvalues_sorted_indices = np.argsort(eigvals)[::-1]
    eigenvalues_sorted = eigvals[eigenvalues_sorted_indices]
    eigenvectors_sorted = eigvecs[:, eigenvalues_sorted_indices]
    # Select the top k eigenvectors
    k = 2
    eigenvectors_subset = eigenvectors_sorted[:, 0:k] #instead of just :k
    print(eigenvectors_subset)
    transformed_X = np.dot(eigenvectors_subset.T, X.T).T
    print(transformed_X.shape)
    return transformed_X
```

```python
new_X = PCA(2, X, model_dim)
```

```python
%matplotlib qt 
plt.scatter(new_X[:,0], new_X[:,1])
for label, x, y in zip(labels, new_X[:, 0], new_X[:, 1]):
    plt.annotate(label, xy = (x,y), xytext = (0, 0), textcoords = "offset points", size = 10)
```
